```python

# Cambio de valor en la misma variable
my_var = 30
print("Nuevo valor de my_var:", my_var)

### Operadores aritméticos

# Suma
suma = 10 + 5
print("Suma:", suma)

# Resta
resta = 10 - 5
print("Resta:", resta)

# Multiplicación
multiplicacion = 10 * 5
print("Multiplicación:", multiplicacion)

# División
division = 10 / 3
print("División:", division)

# División entera
division_entera = 10 // 3
print("División entera:", division_entera)

# Módulo (resto de la división)
modulo = 10 % 3
print("Módulo:", modulo)

# Potenciación
potencia = 2 ** 3
print("Potencia:", potencia)

# Precedencia de operadores (usar paréntesis)
resultado = 10 + 5 * 2
print("Resultado sin paréntesis:", resultado)
resultado = (10 + 5) * 2
print("Resultado con paréntesis:", resultado)

### Uso de módulos math y cmath
import math

# Raíz cuadrada
raiz = math.sqrt(25)
print("Raíz cuadrada de 25:", raiz)

# Entero más grande menor o igual que x
floor_valor = math.floor(3.7)
print("Floor de 3.7:", floor_valor)

# Logaritmo natural de x + 1
logaritmo = math.log1p(9)
print("Logaritmo natural de 9+1:", logaritmo)

# Para números complejos se usa cmath
import cmath
num_complejo = 1 + 2j
print("Módulo del número complejo:", abs(num_complejo))
```
